# 1. Importamos el conector (Framework Connector)
# Google - Gemma 4

```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA
import os
```

```python
llm = ChatNVIDIA(model="google/gemma-4-31b-it")
```

# 2. Configuramos el modelo (La "Capa de Abstracción Alta")
# Esto reemplaza la conexión al 'llm_client' de NVIDIA

```python
prompt = "Hola Gemma, explícame brevemente quien es mil mascaras?"

print(f"--- Consultando a: {llm.model} ---\n")

try:
    # El streaming permite ver la respuesta fluida como en el curso
    for chunk in llm.stream(prompt):
        print(chunk.content, end="", flush=True)
except Exception as e:
    print(f"\n[ERROR]: No se pudo conectar con el modelo. Detalle: {e}")
```
